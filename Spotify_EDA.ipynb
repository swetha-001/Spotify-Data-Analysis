Most popular song across all platforms.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
popularity_cols = [
    'Spotify Popularity', 'Spotify Streams', 'YouTube Views', 'YouTube Likes',
    'TikTok Posts', 'TikTok Likes', 'TikTok Views', 'AirPlay Spins',
    'SiriusXM Spins', 'Shazam Counts', 'Pandora Streams', 'Soundcloud Streams'
]
```

```python
for col in popularity_cols:
    if data[col].dtype == 'object':
        data[col] = data[col].str.replace(',', '').astype(float)
        data[col] = data[col].replace(np.nan,0)
        data[col] = data[col].astype('int')
```

```python
popularity_data = data[popularity_cols]
```

```python
# Normalize the columns using Min-Max Scaling (0 to 1)
scaler = MinMaxScaler()
normalized = scaler.fit_transform(popularity_data)
```

```python
# Create a new DataFrame of normalized scores
normalized_df = pd.DataFrame(normalized, columns=popularity_cols)
```

```python
# Calculate a combined popularity score 
normalized_df['Combined Score'] = normalized_df.sum(axis=1)
```

```python
normalized_df.head(10)
```

```python
# Add back the song and artist names
data_with_scores = data.copy()
data_with_scores['Combined Score'] = normalized_df['Combined Score']
```

```python
most_popular = data_with_scores.loc[data_with_scores['Combined Score'].idxmax()]
```

```python
print("Most Popular Track Across All Platforms:")
print(f"Track Name: {most_popular['Track']}")
print(f"Artist: {most_popular['Artist']}")
print(f"Combined Popularity Score: {most_popular['Combined Score']:.3f}")
```

```python
top10_songs=data_with_scores.sort_values(by='Combined Score',ascending=False).head(10)
```

```python
sns.barplot(data=top10_songs,x='Combined Score',y='Track',hue='Artist',palette='Set1')
plt.title('Top 10 songs across all platforms')
plt.legend(bbox_to_anchor=(1,1),loc=2)
plt.show()
```
